```
iris = pd.read_csv(url)
iris.head()
```

```
len(iris)
```

```
for colname in iris:
    print(colname, iris[colname].dtype)
```

```
iris.variety = iris.variety.apply(str)
```

```
for colname in iris:
    print(colname, iris[colname].dtype)
```

```
from sklearn import tree
```

```
X = iris.iloc[:, :4].to_numpy()
Y = iris.iloc[:, 4].to_numpy()
print(X[:5])
print(Y[:5])
```

```
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```
tree.plot_tree(clf)
```

```
import graphviz as gv
```

```
gv.Source(tree.export_graphviz(clf, out_file=None))
```

```
clf.predict([[5.1, 3.5, 1.4, 0.2]])
```
